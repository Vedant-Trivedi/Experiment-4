```python
A={1,2,3,4}
B={3,4,5,6}
print('Union:',A.union(B))
print('Intersection:',A.intersection(B))
print('Difference A-B:',A.difference(B))
print('Difference B-A:',B.difference(A))
print('Symmetric Difference:',A.symmetric_difference(B))
```

```python
A={1,2,3,4}
B={3,4,5,6}
print('Union:',A|B)
print('Intersection:',A&B)
print('Difference A-B:',A-B)
print('Difference B-A:',B-A)
print('Symmetric Difference:',A^B)
```

```python
'''Problem Statement 1: Unique Event Participants
A college event has a list of students who registered multiple times by mistake
Write a Program using set to store participant names and display only unique participants'''

participants=["Alex","Maya","Daniel","Sofia","Liam","Alex"]
uniqueparticipants= set(participants)
print('Unique Participants:',uniqueparticipants)
```

```python
'''Problem Statement 2: Intersection'''
Student1={'physics','python','chemistry'}
Student2={'biology','chemistry','python'}
Student3={'biology','maths','python'}
print('Common Subject for all Students:',Student1 & Student2 & Student3)
```

```python
'''Problem Statement 3: Intersection and Symmettric Difference'''
CricketCl={"Alex","Maya","Daniel","Sofia","Liam"}
FootballCl={"Alex","Sofia","Liam","Olivia","Oliver"}
print('Students who play both Cricket and Football:',CricketCl & FootballCl)
print('Students who play either Cricket or Football but not both',CricketCl ^ FootballCl)
```

```python
'''Problem Statement 4: Absent Students'''
AllStudents={'Alex','Maya','Daniel','Sofia','Liam','Olivia','Oliver'}
PresentStudents={'Alex','Maya','Daniel','Sofia','Liam'}
print('Absent Students:',AllStudents - PresentStudents)
```

```python
'''Problem Statement 5: Remove Discontinued Subjects using Discard/Remove'''
course={"ETC101","ETC102","ETC103"}
course.remove("ETC102")
print(course)
```

```python
'''Problem Statement 5: Remove Discontinued Subjects using Discard/Remove'''
course={"ETC101","ETC102","ETC103"}
course.discard("ETC102")
print(course)
```
